### 6. Split Dataset

### 7. Baseline Model

```python
y_central = y_train.mode()
y_base = np.full(y_train.shape[0], y_central)
```

```python
from sklearn.metrics import accuracy_score, precision_recall_curve, confusion_matrix, classification_report, roc_auc_score, average_precision_score

train_perf = accuracy_score(y_train, y_base)
valid_perf = accuracy_score(y_val, np.full(y_val.shape, y_val.mode()))
test_perf = accuracy_score(y_test, np.full(y_test.shape, y_test.mode()))

print(f"""Training accuracy : {train_perf*100:.2f}%
Testing accuracy : {test_perf*100:.2f}%
Validation accuracy : {valid_perf*100:.2f}%""")
```

```python
print(classification_report(y_train, y_base))
```

### 8. Train Machine Learning Model

```python
def plot_pr_curve(recall, precision, average_precision):
  plt.step(recall, precision, color='b', alpha=0.2, where='post')
  plt.fill_between(recall, precision, step='post', alpha=0.2, color='b')
  plt.xlabel('Recall')
  plt.ylabel('Precision')
  plt.ylim([0.0, 1.05])
  plt.xlim([0.0, 1.0])
  plt.title('2-class Precision-Recall curve: AP={0:0.2f}'.format(average_precision))
  plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.pipeline import Pipeline

pipeline = Pipeline([
    ('underSampler', RandomUnderSampler(random_state=42)),
    ('classifier', LogisticRegression(random_state=42))
])

pipeline.fit(X_train, y_train)
probs = pipeline.predict_proba(X_train)
predicted = pipeline.predict(X_train)

roc_auc_score(y_train, probs[:,1])
```

```python
probs_test = pipeline.predict_proba(X_test)
predicted_test = pipeline.predict(X_test)

roc_auc_score(y_test, probs_test[:,1])
```

```python
print(classification_report(y_train, predicted))
print(accuracy_score(y_test, predicted_test))
```

```python
precision, recall, _ = precision_recall_curve(y_train, predicted)
plt.plot(recall, precision)
plt.show()
```

```python
average_precision = average_precision_score(y_train, predicted)
precision, recall, _ = precision_recall_curve(y_train, predicted)
plot_pr_curve(recall, precision, average_precision)
```
