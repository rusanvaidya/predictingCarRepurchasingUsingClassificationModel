### 5. Prepare Data

```python
df_cleaned = df.copy()
```

```python
#  Drop the identifier
df_cleaned.drop(['ID'], axis=1, inplace=True)
```

```python
# Filling Missing values
df_cleaned.gender.fillna('Not Specified', inplace=True)
df_cleaned.age_band.fillna('Not Specified', inplace=True)
```

```python
df_cleaned.duplicated().sum() #check duplicates
```

```python
#Drop duplicates
df_cleaned.drop_duplicates(inplace=True)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Preprocessor for SVC
preprocessor = ColumnTransformer(
    transformers = [
        ('ord', StandardScaler(), ord_cols),
        ('cat', OneHotEncoder(), cat_cols)
    ]
)
```

### 6. Split Dataset

```python
y = df_cleaned.pop(target)
X = df_cleaned
```

```python
from sklearn.model_selection import train_test_split

X_data, X_test, y_data, y_test = train_test_split(X, y, stratify=y, random_state=42, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_data, y_data, stratify=y_data, random_state=42, test_size=0.2)
```

### 7. Baseline Model

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SVMSMOTE
from sklearn.svm import SVC

# pipeline for baseline model SVC
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('sampling', SVMSMOTE(random_state=42)),
    ('model', SVC(random_state=42))
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

train_perf = classification_report(y_train, pipeline.predict(X_train))
val_perf = pipeline.score(X_val, y_val)
test_perf = pipeline.score(X_test, y_test)
```

```python
print(train_perf)
print(f"Validation Accuracy: {val_perf:.2%},\
      Test Accuracy: {test_perf:.2%}")
```

### 8. Train Machine Learning Model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline as PIPELINE

# create pipeline for decision tree
model = PIPELINE([
    ('preprocessing', preprocessor),
    ('model', DecisionTreeClassifier(max_depth=10))
])
```

```python
model.fit(X_train, y_train)
```

### 9. Access Model Performance

```python
y_preds = model.predict(X_train)
```

```python
# Display performance on training
print(classification_report(y_train, y_preds))
```

```python
# Display performance on unseen data for max_depth=10
print(f"Validation Accuracy: {model.score(X_val, y_val):.2%}%,\
      Test Accuracy: {model.score(X_test, y_test):.2%}%")
```

```python
# model for pipeline max_depth=20
model_1 = PIPELINE([
    ('preprocessing', preprocessor),
    ('model', DecisionTreeClassifier(max_depth=20))
])
```

```python
model_1.fit(X_train,y_train)
```

```python
# Display performance on training for max_depth=20
print(classification_report(y_train, model_1.predict(X_train)))
```

```python
# Display performance on unseen data for max_depth=20
print(f"Validation Accuracy: {model_1.score(X_val, y_val):.2%}%,\
      Test Accuracy: {model_1.score(X_test, y_test):.2%}%")
```

```python
from sklearn.metrics import PrecisionRecallDisplay, roc_curve, roc_auc_score, make_scorer
```

```python
# Plot precision recall curve
display = PrecisionRecallDisplay.from_estimator(pipeline, X_val, y_val, name='SVC')
display1 = PrecisionRecallDisplay.from_estimator(model, X_val, y_val, ax=display.ax_, name='DecisionTreeClassifier max_depth=10')
display2 = PrecisionRecallDisplay.from_estimator(model_1, X_val, y_val, ax=display.ax_, name='DecisionTreeClassifier max_depth=20')
display.ax_.set_title('Precision-Recall curve')
plt.show()
```

```python
# probabilities for each model
y_probs = pipeline.decision_function(X_test)
y_probs1 = model.predict_proba(X_test)[:, 1]
y_probs2 = model_1.predict_proba(X_test)[:, 1]
```

```python
# ROC curve points
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
fpr1, tpr1, thresholds1 = roc_curve(y_test, y_probs1)
fpr2, tpr2, thresholds2 = roc_curve(y_test, y_probs2)
```

```python
# AUC scores
auc = roc_auc_score(y_test, y_probs)
auc1 = roc_auc_score(y_test, y_probs1)
auc2 = roc_auc_score(y_test, y_probs2)
```

```python
# Plot ROC curve
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, color='red', label=f'SVC (area = {auc:.4f})')
plt.plot(fpr1, tpr1, color='blue', label=f'DecisionTree max_depth=10 (area = {auc1:.4f})')
plt.plot(fpr2, tpr2, color='green', label=f'DecisionTree max_depth=20 (area = {auc2:.4f})')
plt.plot([0, 1], [0, 1], color='black', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
